## Find the five publishers with the highest total global sales. 
## Note: You will need to calculate the total global sales for each publisher to do this.

```python
df[['Publisher', 'Global_Sales']].groupby(['Publisher']).sum().sort_values(by='Global_Sales', ascending=False).head(n=5)
```

## Create a new column in the data frame that calculates the percentage of global sales from North America. 
## Display the first five rows of the new data frame.

```python
df["Percentage_NASales_GlobalSales"] = (df["NA_Sales"] / df["Global_Sales"]*100)
df.head(5)
```

## Find the number NaN entries (missing data values) in each column.

```python
def describe_nan(df):
    return pd.DataFrame([(i, df[df[i].isna()].shape[0],df[df[i].isna()].shape[0]/df.shape[0]) for i in df.columns], columns=['column', 'nan_counts', 'nan_rate'])

describe_nan(df)
```

## Try to calculate the median user score of all the video games. 
## You will likely run into an error because some of the user score entries are a non-numerical string that cannot be converted to a float. 

```python
df.head(5)
```

```python
## Find and replace this string with NaN
import numpy as np
median_df = df
median_df = median_df.replace('tbd', np.nan)
median_df.head(122)
```

```python
## Then calculate the median. 
median_value = median_df["User_Score"].median()
print(median_value)
```

```python
## Then, replace all NaN entries in the user score column with the median value.
median_df['User_Score'].fillna(value=median_df['User_Score'].median(), inplace=True)
median_df.head(122)
```
